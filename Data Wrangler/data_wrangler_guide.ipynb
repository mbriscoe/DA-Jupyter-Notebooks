## 1. What is Data Wrangler?

Data Wrangler is a VS Code extension that provides:
- **Visual data exploration**: Interactive data viewer with sorting and filtering
- **No-code transformations**: Point-and-click data cleaning operations
- **Code generation**: Automatically generates Python/Pandas code for your transformations
- **Data profiling**: Quick statistical summaries and data quality insights
- **Export capabilities**: Apply transformations and export cleaned data

It's perfect for both beginners and experienced data analysts who want to speed up data cleaning workflows.

## 2. Installing Data Wrangler

To install Data Wrangler:
1. Open VS Code Extensions (Ctrl+Shift+X)
2. Search for "Data Wrangler"
3. Click Install on the Microsoft Data Wrangler extension
4. Reload VS Code if prompted

**Prerequisites:**
- Python extension for VS Code
- Pandas library installed (`pip install pandas`)

## 7. Column Transformations

Let's look at transformations you can apply:

```python
# Example transformations you can do in Data Wrangler:

# 1. Create new column (Revenue per item)
df['Revenue_Per_Item'] = df['Sales'] / df['Quantity']

# 2. Convert data types
df['Date'] = pd.to_datetime(df['Date'])

# 3. Extract date components
df['Day_of_Week'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month

# 4. String operations
df['Product_Upper'] = df['Product'].str.upper()

print("Transformed DataFrame:")
df.head()
```

## 8. Aggregation and Grouping

Data Wrangler supports grouping operations:

```python
# In Data Wrangler, you can:
# 1. Group by one or more columns
# 2. Select aggregation functions (sum, mean, count, min, max)
# 3. Preview aggregated results

# Example: Sales by Category
category_sales = df.groupby('Category').agg({
    'Sales': ['sum', 'mean', 'count'],
    'Quantity': 'sum',
    'Returns': 'sum'
})

print("Sales by Category:")
category_sales
```

## 10. Exporting Transformed Data

After applying transformations in Data Wrangler:

### Export Options:
1. **Export to Notebook**: Inserts generated code into a new cell
2. **Copy Code**: Copies transformation code to clipboard
3. **Export to CSV**: Saves transformed data as CSV file
4. **Create New Variable**: Assigns result to a new DataFrame variable

### Code Generation:
- Data Wrangler generates clean, readable Pandas code
- Code includes all transformations in order
- Can be edited and reused
- Maintains reproducibility

```python
# Example of exported code from Data Wrangler
# This is what Data Wrangler might generate:

def clean_data(df):
    """
    Data cleaning function generated by Data Wrangler
    """
    # Filter rows
    df = df[df['Sales'] > 50]
    
    # Sort by Sales descending
    df = df.sort_values(by='Sales', ascending=False)
    
    # Create new column
    df['Sales_Category'] = df['Sales'].apply(
        lambda x: 'High' if x > 500 else 'Low'
    )
    
    return df

# Apply transformations
cleaned_df = clean_data(df.copy())
print("Cleaned DataFrame:")
cleaned_df
```

## 15. Troubleshooting Tips

### Common Issues:

**Issue**: Data Wrangler won't open
- **Solution**: Ensure DataFrame variable is created and visible in Variables pane
- Check that Pandas is installed
- Restart the kernel

**Issue**: Transformations not working
- **Solution**: Check data types of columns
- Ensure no null values in key columns
- Try operation on smaller dataset first

**Issue**: Performance is slow
- **Solution**: Sample large datasets first
- Close other applications
- Consider filtering data before opening Data Wrangler

**Issue**: Generated code has errors
- **Solution**: Check for special characters in column names
- Verify data types match operation requirements
- Test code on sample data first
